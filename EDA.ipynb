```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder
from pandas.api.types import is_string_dtype
```

```python
df1=pd.read_csv(r'C:\Users\aliss\DS-Project\WINEDATA_CLEAN.csv')
df=df1.copy()
```

Wine is usually in bars...

```python
mycolors = ['#CB4335', '#E74C3C','#EC7063','#F1948A','#F5B7B1']
df["Price Range"].value_counts().plot(kind='bar', color = mycolors)
plt.ylabel('Count')
plt.xlabel('Price')
```

But who doesn't like pies?

```python
df["Price Range"].value_counts().plot(kind='pie',colors=mycolors)
plt.title("Price Range")
plt.ylabel('')
```

What affects wine prices?


```python

fig = plt.figure(figsize=(25,15))

colors=['#21618C','#ECF0F1', '#E74C3C']
df["Country"].value_counts().plot(kind='bar', color = colors)
```

```python
plt.scatter(df['Type & Color'], df['ABV'])
plt.xlabel('Type & Color')
plt.ylabel('ABV')
plt.show()
```

```python
#vintage&price
print(df['Price'].corr(df['Vintage']))
print(df['Price'].corr(df['ABV']))
print(df['Price'].corr(df['Vintage']))
```
