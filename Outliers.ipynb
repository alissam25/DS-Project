```python
#OUTLIERS
#super high prices- what do they have in common?
df["Regular Price"].describe()
Q1 = np.percentile(df["Regular Price"], 25)
Q3 = np.percentile(df["Regular Price"], 75)
IQR = Q3 - Q1
Fare_outlier_rows = df[(df["Regular Price"] < Q1 - 1.5*IQR) | (df["Regular Price"] > Q3 + 1.5*IQR )]
Fare_outlier_rows.sort_values("Regular Price")
Fare_outlier_rows["Regular Price"].min()

sns.boxplot(Fare_outlier_rows["Regular Price"])
```
